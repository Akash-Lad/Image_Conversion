# Image Conversion

By default the images in OpenCV are read in BGR channel. Hence, it has to be converted to RGB.

Converting the image from BGR to RGB channel. To view image in grey-scale, use cmap='gray' or image dimension can be reduced to 127 out 255 channels.

```python
new_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
plt.imshow(new_pic)
```

```python
img_gray = cv2.cvtColor(pic,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
new_pic.shape
```

```python
# re-sizing the image

height = int((new_pic.shape[0])/2)
width = int((new_pic.shape[1])/4)

img = cv2.resize(new_pic,(height,width))
plt.imshow(img,cmap='gray')
```

# Flipping Images

```python
flip_img = cv2.flip(img,0)
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(img,1)
plt.imshow(flip_img)
```
